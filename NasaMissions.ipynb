```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pyodbc
import pandas as pd
import requests
```

```python
port=5432
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://nasa.gov/missions')
```

```python
soup=BeautifulSoup(driver.page_source, 'html.parser')
listbody=soup.find('div', class_='static-landing-page__body')
```

```python
missionsp=listbody.find_all('p')
```

```python
missionlinks=[]
for p in missionsp:
    missions=p.find_all('a')
    for a in missions:
        try:
            mission=a.text
            link=a['href']
            if 'http://' in link:
                missionlinks.append([mission,link])
        except:
            pass
```

```python
missionpagecontent=[]
for mission in missionlinks:
    try:
        missionurldata=requests.get(mission[1]).text
    except:
        missionurldata=''
    missionpagecontent.append([mission[0],mission[1],missionurldata])
nasamissionsdf=pd.DataFrame(missionpagecontent)
```
